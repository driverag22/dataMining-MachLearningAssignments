```python
def f(x):
    return (1/2) *(x[0] ** 4) - x[0] * x[1] + (x[1] ** 2) + x[1] * x[2] + (x[2] ** 2)

def f_gradient_at(x):
    return [2*(x[0]**3) - x[1], -x[0] + 2*x[1] + x[2], x[1] + 2*x[2]]

def gradiant_eq_zero(i, x):
    if i == 0: # first component of gradient eq. to zero, solve for x_1 = x[0]
        return (x[1]/2)**(1/3)
    elif i == 1: # sec. component eq. to zero, solve for x_2 = x[1]
        return (x[0]-x[2])/2
    elif i == 2: # ...
        return -x[1]/2
```

```python
## 2a
# Function is convex (or at least hessian is PSD at x_0), so we just find values s.t. component i of 
# the gradiant equals 0
x_0=[2,3,4]

def argmin_x1(x): # mind the index shift
    return gradiant_eq_zero(0, x)

def argmin_x2(x):
    return gradiant_eq_zero(1, x)

def argmin_x3(x):
    return gradiant_eq_zero(2, x)

def argmin(i, x):
    match i:
        case 0:
            return argmin_x1(x)
        case 1:
            return argmin_x2(x)
        case 2:
            return argmin_x3(x)
        case _:
            return -1

for i in range(0,3):
    print("i = " + str(i) + ": " + str(argmin(i, x_0)) )
    print()

## quick test
for t in range(-100, 100):
    if ( f( [t, x_0[1], x_0[2]] ) < f( [argmin(0, x_0), x_0[1], x_0[2]] ) ):
        print("not min.")
    if ( f( [x_0[0], t, x_0[2]] ) < f( [x_0[0], argmin(1, x_0), x_0[2]] ) ):
        print("not min.")
    if ( f( [x_0[0], x_0[1], t] ) < f( [x_0[0], x_0[1], argmin(2, x_0)] ) ):
        print("not min.")
```

```python
## 2b
x_0=[1,20,5]
def coordinate_descent(f, argmin, x, max_iter=100):

    print(str(0) + ":")
    print(x)
    print(f(x))
    print()

    for j in range(1, max_iter+1):
        # use temp to save values of argmin we find
        temp = x
        print(str(j) + ":")
        for i in range(0,3):
            temp[i] = argmin(i, x) 
        x = temp
        print(x)
        print(f(x))
        print()

coordinate_descent(f, argmin, x_0, 100)
```
